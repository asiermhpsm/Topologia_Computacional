## Circulo: alfa-complejo vs Vietoris-Rips

```python
num_points = 10
radio=1
max_error=0.10

# Generar ángulos aleatorios
angles = np.random.uniform(0, 2 * np.pi, num_points)

# Generar radios con un poco de ruido
true_radii = np.ones(num_points) * radio
noisy_radii = true_radii + np.random.uniform(-max_error, max_error, num_points)

# Convertir coordenadas polares a cartesianas
x = noisy_radii * np.cos(angles)
y = noisy_radii * np.sin(angles)

points=np.array([[x[i], y[i]] for i in range(num_points)])
```

```python
ac = AlphaComplex(points)
vr = VietorisRips(points)
```

```python
ac.representaPuntos()
```

```python
ac.diagramaPersistencia()
vr.diagramaPersistencia()
```
